```python
import numpy as np
import matplotlib.pyplot as plt
%load_ext csoundmagics
```

```python
cs = ICsound()
```

```python
def plotCurve(ax, x, y):
    minVal = min(x.min(), y.min())
    maxVal = max(x.max(), y.max())
    ax.set_xlim(minVal, maxVal)
    ax.set_ylim(minVal, maxVal)
    ax.set_aspect('equal')
    ax.plot(x, y)
```

```python
cs.makeTable(10, 2048, -9, 5, 4, 0)
cs.makeTable(11, 2048, -9, 6, 7.2, 90)
cs.plotTable(10)
cs.plotTable(11, reuse=True)
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot(1, 1, 1)
x = cs.table(11)
y = cs.table(10)
plotCurve(ax, x, y)
```
